```python
import logging

logger = logging.getLogger("python_coding_agent")
logger.setLevel(logging.DEBUG)
logger.handlers.clear()

handler = logging.StreamHandler()
handler.setFormatter(logging.Formatter("%(levelname)s:%(name)s:%(message)s"))
logger.addHandler(handler)
```

```python
from python_coding_agent import CodingAgent


agent = CodingAgent(
    model="gpt-5-nano",
    max_turns=3,
)
# print(agent)

task = "Create a python function which return the maximum number from a given list of numbers"
response = agent.call(task)

print("\n######## RESULT ########")
print(response.output_text)
```
