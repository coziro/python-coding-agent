```python
from openai import OpenAI
from python_coding_agent import CodingAgent

model = "gpt-5-nano"
client = OpenAI()

agent = CodingAgent(
    client=client,
    model=model,
    max_turns=3,
)
print(agent)

task = "Create a python function which return the maximum number from a given list of numbers"
response = agent.call(task)

print("\n######## RESULT ########")
print(response.output_text)
```
